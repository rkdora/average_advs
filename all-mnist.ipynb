```python
import numpy as np
import matplotlib.pyplot as plt
from keras.datasets import mnist
```

```python
# データの読み込み
(X_train, y_train), (X_test, y_test) = mnist.load_data()
```

```python
n = 10
```

```python
stack_list = []
for _ in range(10):
    stack_list.append([])
    
stack_list
```

```python
fill_list = []
for i, y in enumerate(y_test):
    if (not y in fill_list) and len(stack_list[y]) < n:
        stack_list[y].append(X_test[i].flatten())
    else:
        fill_list.append(y)
```

```python
for i, stacks in enumerate(stack_list):
    print(i, len(stacks))
```

```python
avg_list = []
for stacks in stack_list:
    avg = np.mean(stacks, axis=0)
    avg_list.append(avg)
```

```python
for avg in avg_list:
    plt.imshow(avg.reshape(28,28), 'gray')
    plt.show()
```
